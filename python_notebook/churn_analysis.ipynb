```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('../dataset/WA_Fn-UseC_-Telco-Customer-Churn.csv')
```

```python
df.head()
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
df['TotalCharges'] = pd.to_numeric(
    df['TotalCharges'],
    errors='coerce'
)
```

```python
df['TotalCharges'].fillna(
    df['TotalCharges'].median(),
    inplace=True
)
```

```python
df.duplicated().sum()
```

```python
df.drop_duplicates(inplace=True)
```

```python
df.shape
```

```python
sns.countplot(x='Churn', data=df)
plt.show()
```

```python
churn_rate = (
    df['Churn'].value_counts()['Yes']
    / len(df)
) * 100

print(churn_rate)
```

```python
sns.countplot(
    x='Contract',
    hue='Churn',
    data=df
)

plt.xticks(rotation=45)
plt.show()
```

```python
sns.countplot(
    x='PaymentMethod',
    hue='Churn',
    data=df
)

plt.xticks(rotation=90)
plt.show()
```

```python
sns.countplot(
    x='InternetService',
    hue='Churn',
    data=df
)

plt.show()
```

```python
sns.boxplot(
    x='Churn',
    y='MonthlyCharges',
    data=df
)

plt.show()
```

```python
sns.histplot(df['tenure'], bins=30)
plt.show()
```

```python
df['CustomerSegment'] = np.where(
    (df['tenure'] < 12)
    &
    (df['MonthlyCharges'] > 70),

    'High Risk',

    'Normal'
)
```

```python
df['TenureGroup'] = pd.cut(
    df['tenure'],

    bins=[0,12,24,48,60,72],

    labels=[
        '0-1 Year',
        '1-2 Years',
        '2-4 Years',
        '4-5 Years',
        '5-6 Years'
    ]
)
```

```python
sns.countplot(
    x='TenureGroup',
    hue='Churn',
    data=df
)

plt.xticks(rotation=45)
plt.show()
```

```python
df.to_csv(
    '../Dataset/cleaned_churn_data.csv',
    index=False
)
```
